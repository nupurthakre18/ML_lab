```python
data_series = pd.Series(df1['Age']);

# Calculate Q1 and Q3
q1 = data_series.quantile(0.25)
q3 = data_series.quantile(0.75)

# Calculate the IQR
iqr = q3 - q1
lb = q1-1.5*iqr
ub = q3+1.5*iqr
print(iqr,lb,ub)
```

```python
columns_to_plot = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
plt.figure(figsize=(10, 6))

box_plots = plt.boxplot([df[col] for col in columns_to_plot],
                        labels=columns_to_plot,
                        patch_artist=True)
for box in box_plots['boxes']:
    box.set_facecolor('green')

plt.title("Boxplot of Selected Columns")
plt.ylabel('Values')
plt.show()
```

```python
df_cleaned = df[(df['Age'] >= lb) & (df['Age'] <= ub)]
df_cleaned
```

```python
columns_to_plot = ['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

plt.figure(figsize=(12, 10))
box_plots = plt.boxplot([df_cleaned[col] for col in columns_to_plot],
                        labels=columns_to_plot,
                        patch_artist=True)
for box in box_plots['boxes']:
    box.set_facecolor('green')

plt.title("Boxplot of Selected Columns")
plt.ylabel('Values')
plt.show()
```

```python
plt.hist(df_cleaned['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df_cleaned['Age'], df_cleaned['Fare'], color='coral', alpha=0.7)
plt.title('Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
import matplotlib.pyplot as plt
pclass_counts = df_cleaned['Pclass'].value_counts()
plt.bar(pclass_counts.index, pclass_counts.values, color=['lightgreen', 'lightblue', 'lightcoral'])
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
survival_counts = df_cleaned['Survived'].value_counts()
plt.pie(survival_counts.values, labels=survival_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
plt.title('Survival Status')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
numeric_df = df_cleaned.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
